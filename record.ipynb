```python
sr = 44100

wav = wave.open("test.wav", "rb")
nframes = wav.getnframes()
framerate = wav.getframerate()
AUDIO_LEN = nframes / framerate
print(AUDIO_LEN)
```

```python
from faster_whisper import WhisperModel

model_size = "tiny"
model = WhisperModel(model_size, device="cpu", compute_type="int8")
```

```python
segments, info = model.transcribe("test.wav", beam_size=5)

print("Detected language '%s' with probability %f" % (info.language, info.language_probability))

for segment in segments:
    print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))
```

```python
import time

def test_model(model: WhisperModel, **kwargs):
    N = 10
    times = []
    for _ in range(N):
        t = time.time()
        segments, info = model.transcribe("test.wav", **kwargs) 
        times.append(time.time() - t)
        for segment in segments:
            print("[%.2fs -> %.2fs] %s" % (segment.start, segment.end, segment.text))

    # average time in seconds
    print(f"Mean transcription time: {np.mean(times):.2f} s")
    print(f"Mean RTF: {np.mean(times) / AUDIO_LEN:.2f}")
```

```python
test_model(model)
```

```python
test_model(model, beam_size=5, language="en", best_of=3, word_timestamps=False, without_timestamps=True)
```
